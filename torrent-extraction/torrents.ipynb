```python
import json
import os

```

```python

folder_path = "data\Torrent"
json_files = [f for f in os.listdir(folder_path) if f.endswith('.json')]

print(json_files)
```

```python
results = []

for file_name in json_files:
    with open(os.path.join(folder_path, file_name)) as f:
        data = json.load(f)
        results.extend(data["results"])

print(results)
```

```python
data_dict = {}
count = 0
for file_name in json_files:
    with open(os.path.join(folder_path, file_name)) as f:
        data = json.load(f)
        for entry in data["results"]:
            count = count + 1
            release_name = entry["release_name"]
            size = int(entry["size"])
            if release_name not in data_dict:
                data_dict[release_name] = entry
            elif size < int(data_dict[release_name]["size"]):
                data_dict[release_name] = entry

unique_entries = list(data_dict.values())

print(unique_entries)
```

```python
import libtorrent as lt

# create session
ses = lt.session()
ses.listen_on(6881, 6891)

# add torrent file to session
info = lt.torrent_info()
h = ses.add_torrent({'ti': info, 'save_path': 'data\Torrent\downloads'})

# download torrent
print('Downloading:', h.name())
while (not h.is_seed()):
    s = h.status()
    print(s.progress * 100)
```
